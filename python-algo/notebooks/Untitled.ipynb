```python
import pandas as pd

# Load JSON files
df_activities = pd.read_json("activities.json")
df_likes = pd.read_json("likes.json")
df_views = pd.read_json("views.json")
df_comments = pd.read_json("comments.json")

# Ensure captions are strings
df_activities["caption"] = df_activities["caption"].fillna("")

# Quick view of the data
df_activities.head()
df_likes.head()
df_views.head()
df_comments.head()
```

```python
print("Activities:", len(df_activities))
print("Likes:", len(df_likes))
print("Views:", len(df_views))
print("Comments:", len(df_comments))
```

```python
# Create engagement scores for each activity
engagement_scores = {}

# Add points for views (1 point each)
for _, row in df_views.iterrows():
    activity_id = row['activity_id']
    engagement_scores[activity_id] = engagement_scores.get(activity_id, 0) + 1

# Add points for likes (2 points each)
for _, row in df_likes.iterrows():
    activity_id = row['activity_id']
    engagement_scores[activity_id] = engagement_scores.get(activity_id, 0) + 2

# Add points for comments (3 points each)
for _, row in df_comments.iterrows():
    activity_id = row['activity_id']
    engagement_scores[activity_id] = engagement_scores.get(activity_id, 0) + 3

print(f"User has engaged with {len(engagement_scores)} unique activities")
print(f"Example scores: {list(engagement_scores.items())[:5]}")
```

```python
# Get list of activity IDs the user has already engaged with
engaged_activity_ids = set(engagement_scores.keys())

# Get all activity IDs from the system
all_activity_ids = set(df_activities['id'])

# Find activities the user hasn't seen
unseen_activities = all_activity_ids - engaged_activity_ids

print(f"Activities user has NOT seen yet: {len(unseen_activities)}")
print(f"These are the candidates for recommendation")
```

```python
# Filter unseen activities
recommendations = df_activities[df_activities['id'].isin(unseen_activities)].copy()

# Sort by creation date (newest first)
recommendations = recommendations.sort_values('created_at', ascending=False)

# Show top 10 recommendations
print(f"Top 10 Recommended Activities:\n")
print(recommendations[['id', 'owner_id', 'created_at', 'is_public']].head(10))

print(f"\nTotal recommendations available: {len(recommendations)}")
```
